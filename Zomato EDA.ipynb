```python
rating=data_merge.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_Count'})
```

```python
rating
```

# Conclusion
1.When rating is 4.5 to 4.9->Excellent
2.When rating are between 4.0 to 4.4-> Very Good
3.When rating are between 3.5 to 3.9-> Good
4.When rating are between 2.5 to 3.4 ->Average
5.When ratings are between 1.8 to 2.4-> Poor

```python
rating.head()
```

```python
sns.barplot(x='Aggregate rating',y='Rating_Count',hue='Rating color',data=rating)
```

```python
# Mapping the color pallete

# Specify the color palette with corrected shades of green
color_palette = ['white', 'red', 'orange', 'yellow', 'green', 'darkgreen']

sns.barplot(x='Aggregate rating', y='Rating_Count', hue='Rating color', data=rating, palette=color_palette)
```

# Observation
1.Not Rated Count is very high
2.Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot( x='Rating color',data=rating,palette=['white','red','orange','yellow','green','darkgreen'])
plt.show()
```
